```python

import numpy as np
import matplotlib.pyplot as plt

pi=np.pi

r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(5,4.5),dpi=200)
#ax.plot(theta, r)
#ax.set_rmax(2)
#ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
#ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)
offset=-12.3*pi/8


arr1 = plt.arrow(offset, 0.01, 0, 0.7, alpha = 1, width = 0.015,
                 edgecolor = 'black', facecolor = 'black', lw = 2, zorder = 5)
plt.plot([0+offset+pi,1+offset+pi],[0,0.95],color='DarkBlue')
plt.plot([1+offset+pi],[0.95],'o',color='blue')
plt.plot([0+offset,1+offset],[0,0.5],':',color='DarkBlue')
plt.text(-0.1+offset,0.7,r"$\vec{E}(t)$",color='black')
plt.text(1.12+offset+pi,0.5,r"$\chi$",color='DarkBlue')
ax.set_yticks([])
theta=np.arange(0,1,0.01)
ax.plot(theta+offset,np.ones_like(theta)*0.3,color='DarkRed')
lab=plt.text(1.1+offset,0.5,r"$\eta=\mathrm{arctan}\left(-\frac{D}{\tilde{\Omega}}\right)$",color='DarkRed',rotation=0.2,ha='left',va='bottom')
lab.set_rotation(113)
theta2=np.arange(offset,-2.*3.14,-0.01)
ax.plot(theta2,np.ones_like(theta2)*0.15,'-',color='black')
ax.text(0.15,0.2,r'$\tilde{\Omega}$t')
ax.text(1.05+offset+pi,0.91,r'$\vec{\xi}(t)$',color='blue')
#plt.plot([0,1],[0,1])
arr2 = plt.arrow(279*pi/180, 0.8, 2*pi/180, 0, alpha = 1, width = 0.015,
                 edgecolor = 'DarkGreen', facecolor = 'DarkGreen', lw = 0., zorder = 5)

plt.ylim(0,1.1)
#plt.xticks(np.arange(0,2*pi,0.25*pi),labels=[r'0$^\circ$',r'315$^\circ$',r'270$^\circ$',r'225$^\circ$',r'180$^\circ$',r'135$^\circ$',r'90$^\circ$',r'45$^\circ$'])
thetas=np.arange(170,280,1)*pi/180
ax.plot(thetas,np.ones_like(thetas)*0.8,color='DarkGreen')
lab2=plt.text(225*pi/180,0.6,"direction of \n micro-motion",color='DarkGreen',ha='center',va='center',alpha=0.7)
lab2.set_rotation(-45)

plt.savefig("PolarPlot.png",dpi=200,bbox_inches='tight')
plt.show()
```
